```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.patch.set_facecolor('white')  # Цвет фона всей фигуры
ax.set_facecolor('#fffacd')  # Цвет фона только для графика

plt.plot([16, 12, 3, 17, 14], [1, 2, 3, 4, 5], c='#28464d', linestyle='--')
plt.grid(color='#500917',linestyle=':')	# включение отображения сетки


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создание данных
x = np.linspace(0, 90, 50)
y = x**2+6*x+9

fig, ax = plt.subplots()
fig.patch.set_facecolor('white')  # Цвет фона всей фигуры
ax.set_facecolor('#cad5ce')  # Цвет фона только для графика

# Построение графика
plt.title('Квадратичная зависимость y = x**2+6x+9') # заголовок
plt.xlabel('x') # ось абсцисс
plt.ylabel('y') # ось ординат
plt.grid(color='#668572', linestyle=':') # изменение цвета сетки на синий
plt.plot(x, y, c='#61515b', linestyle='--') # построение графика

# Отображение графика
plt.show()
```

```python
# Линейная зависимость
import matplotlib.pyplot as plt
x = np.linspace(0, 20, 50)
y1 = x
 #	Квадратичная зависимость
y2 = [i**2 for i in x]

fig, ax = plt.subplots()
fig.patch.set_facecolor('white')  # Цвет фона всей фигуры
ax.set_facecolor('#fadadd')  # Цвет фона только для графика

#	Построение графика
plt.title('Зависимости: y1 = x, y2 = хл2') # заголовок
plt.xlabel('x',)	# ось абсцисс
plt.ylabel('y1, y2',)	# ось ординат
plt.grid(color='#560319', linestyle='--') # изменение цвета сетки на синий
# Построение графиков с разными цветами
plt.plot(x, y1, color='#b591b5', label='y1 = x')  # линейная зависимость
plt.plot(x, y2, color='#270847', label='y2 = x^2')   # квадратичная зависимость
```

```python
import matplotlib.pyplot as plt  # Импортирует библиотеку matplotlib для создания графиков
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)  # Импортирует инструменты для настройки меток на осях
import numpy as np  # Импортирует библиотеку numpy для работы с массивами данных

x = np.linspace(0, 10, 10)  # Создает массив x из 10 равноотстоящих точек от 0 до 10
y1 = 4 * x  # Вычисляет значения y1 как 4 * x
y2 = [i**2 for i in x]  # Вычисляет значения y2 как квадрат каждого элемента в x (используя list comprehension)

fig, ax = plt.subplots(figsize=(8, 6))  # Создает фигуру и оси графика с размером 8x6 дюймов

ax.set_title('График зависимости: y1=4*x, y2=x**2', fontsize=16)  # Устанавливает заголовок графика
ax.set_xlabel('x', fontsize=14)  # Устанавливает подпись оси x
ax.set_ylabel('y1, y2', fontsize=14)  # Устанавливает подпись оси y
ax.grid(which='major', linewidth=1.2, color='#b591b5')  # Включает основную сетку с заданной толщиной и цветом
ax.grid(which='minor', linestyle='-', color='pink', linewidth=0.5)  # Включает вспомогательную сетку с заданным стилем, цветом и толщиной
ax.scatter(x, y1, c='violet', label='y1 = 4*x')  # Рисует точки для y1 фиолетовым цветом и добавляет метку для легенды
ax.scatter(x, y2, c='turquoise', label='y2 = x**2')  # Рисует точки для y2 бирюзовым цветом и добавляет метку для легенды
ax.legend()  # Отображает легенду на графике

ax.xaxis.set_minor_locator(AutoMinorLocator())  # Автоматически определяет положение вспомогательных делений на оси x
ax.yaxis.set_minor_locator(AutoMinorLocator())  # Автоматически определяет положение вспомогательных делений на оси y

ax.tick_params(which='major', length=10, width=2)  # Настраивает параметры основных делений (длина и толщина)
ax.tick_params(which='minor', length=5, width=1)  # Настраивает параметры вспомогательных делений (длина и толщина)
```
